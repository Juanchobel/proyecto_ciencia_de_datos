```python
import pandas as pd
import numpy as np

dataset = pd.read_csv('housing.csv')

dataset['total_bedrooms'] = dataset['total_bedrooms'].fillna(dataset['total_bedrooms'].median())

dataset_encoded = pd.get_dummies(dataset, columns=['ocean_proximity'], drop_first=True)

X = dataset_encoded.drop('median_house_value', axis=1)
y = dataset_encoded['median_house_value']

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
Xsc = sc.fit_transform(X)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(Xsc, y, test_size=0.2, random_state=5)

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Modelo de Regresión Lineal
modelo_rl = LinearRegression()
modelo_rl.fit(X_train, y_train)
y_pred = modelo_rl.predict(X_test)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(" RMSE:", rmse)
print(" MAE:", mae)
print(" R^2:", r2)
```

```python
import joblib

joblib.dump(modelo_rl, 'linear_modelo.pkl')
print("Modelo Regresión Lineal guardado como linear_modelo.pkl")

#  Guradando el scaler para usarlo en Flask
joblib.dump(sc, 'scaler.pkl')
print("Scaler guardado como scaler.pkl")
```
